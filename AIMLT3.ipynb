```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=n)
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='rating', data=n, order=n['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='country', data=n, order=n['country'].value_counts().index)
plt.title('Distribn of countries')
plt.xlabel('count')
plt.ylabel('counties')
plt.show()
```

```python
#3
sns.boxplot(x=n['release_year'])
```

```python
#4
plt.figure(figsize=(6,4))
corr = n[['release_year']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
